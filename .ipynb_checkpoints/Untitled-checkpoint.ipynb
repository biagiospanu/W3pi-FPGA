```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_hdf('data.h5')
data
```

```python
train_data=data[data['is_train']]
train_data
```

```python
train_signal_data = train_data[train_data['class']==1]
train_signal_data = train_signal_data.to_numpy()
train_base_data = train_data[train_data['class']==0]
train_base_data = train_base_data.to_numpy()
```

```python
train_signal_data.shape
```

```python
n=3
for n in range(3, 63):
    n_bins=20
    min_val=np.min([train_signal_data[n], train_base_data[n]])
    max_val=np.max([train_signal_data[n], train_base_data[n]])
    if min_val>0:
        bin_edges=np.logspace(np.log10(min_val), np.log10(max_val), n_bins)
    else:
        bin_edges=np.linspace(min_val, max_val, n_bins)
    plt.clf()
    plt.hist(train_signal_data[n], label='signal', density=True, bins=bin_edges, log=True)
    plt.hist(train_base_data[n], label='base', density=True, bins=bin_edges, alpha=0.6, log=True)
    plt.xlabel(data.columns[n])
    plt.legend()
    #plt.xscale('log')
    plt.savefig(f'features/{data.columns[n]}.png')
```
